1. Create a SQL table named movie rating with attributes like movie name, genre, and rating
2. Insert records into SQL table through SQL statement 
3. Fetch and view the result
4. Create a dataframe of the result

```python
q1 = """ create table movie_rating
(movie_name varchar(30), genre varchar(20), rating real);"""
```

```python
execute1 = sqlite3.connect(':memory:')
execute1.execute(q1)
execute1.commit()
```

```python
insert2 = [("A Thursday", "Thriller", 8.9),
          ("Love Aaj Kal", "Romantic", 6.5),
          ("Pyar ka Punchnama", "Rom-Com", 7.2),
          ("83", "Documentery", 9),
          ("Action Friday", "Action", 4.2)]

execute1.executemany("Insert into movie_rating values (?,?,?)", insert2)
execute1.commit()
```

```python
resultset = execute1.execute("Select * from movie_rating")
resultset.fetchall()
```

```python
q11 = "select * from movie_rating"
result_df = pd.read_sql(q11, execute1)
result_df
```

```python
import sqlite3
query1 = """ create table customers
(ID integer, customer_name varchar(30), billing_address varchar(50), shipping_address varchar(50));"""
```

```python
exec1 = sqlite3.connect(':memory:')
exec1.execute(query1)
exec1.commit()
```

```python
query2 = exec1.execute("select * from customers")
```

```python
query2.fetchall() #Empty set as no data inserted.
```
