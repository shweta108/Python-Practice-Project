Plot Total Sales Per Month for the year 2011. How has the total sales increased over the months? Which month has the lowest sales?

Plot Total Sales Per Month for the year 2011 in a bar chart. Is bar chart better to visualize than a simple plot?

Plot a pie chart for the year 2010, country wise. Which country contributes the highest and lowest towards sales? Create a pandas series with indexes of the country-wise sales.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("BigMartSalesData.csv")
df.head(2)
```

```python
df.columns
```

Plot Total Sales Per Month for the year 2011. How has the total sales increased over the months? Which month has the lowest sales?

```python
df_2011 = df[df["Year"] == 2011]
df_2011
```

```python
sales_month_2011 = df_2011.groupby("Month").sum()["Amount"]
sales_month_2011
```

```python
type(sales_month_2011)
```

```python
sales_month_2011.index #Index is month
```

```python
sales_month_2011.values #Total sales as values
```

```python
plt.plot(sales_month_2011.index, sales_month_2011.values)
plt.title("Total Monthly Sales for Year 2011")
plt.xlabel("Month in Number")
plt.ylabel("Sales")
plt.show()
```

Plot Total Sales Per Month for the year 2011 in a bar chart. Is bar chart better to visualize than a simple plot?

```python
plt.bar(sales_month_2011.index, sales_month_2011.values)
plt.title("Total Monthly Sales for Year 2011")
plt.xlabel("Month in Number")
plt.ylabel("Sales")
plt.show()
```

Plot a pie chart for the year 2010, country wise. Which country contributes the highest and lowest towards sales? 

```python
df_2010 = df[df["Year"] == 2010]
df_2010
```

```python
sales_country_2010 = df_2010.groupby("Country").sum()["Amount"]
```

```python
#Create a pandas series with indexes of the country-wise sales. Series sorted in asc order
sales_country_2010.sort_values()
```

```python
plt.pie(sales_country_2010.values,labels=sales_country_2010.index)
plt.title("Country-wise Sales in the Year 2010")
plt.show()
```
